# Semantic similarity

## Wordnet

Semantic dictionary of English words, interlinked by semantic relations.

Wornet organizes information in a hierarchy.

Many similarity measures use the hierarchy in some way.

Verbs, nouns, adjectives all have seperate hierarchies.

```python
import nltk
from nltk.corpus import wordnet as wn
```

```python
deer = wn.synset('deer.n.01') #first noun meaning
elk = wn.synset('elk.n.01')
```

```python
deer.path_similarity(elk)
```

# Topic Modelling

A coarse-level analysis of what is in a text collection.

Topic: the subject of a discourse.

Topics are represented as a word distribution.

A document is a mixture of topics.

What is known at topic modelling?

    - Text collection or corpus
    - Number of topics

What is not known?

    - The actual topics
    - Topic distribution for each document
    
Essencially, text clustering problem

    - Documents and words clustered simultaneously
    
Topic modelling approaches

    - Probabilistic Latent Semantic Analysis
    - Latent Dirichlet Allocation(LDA)

## Application of Topic Modelling

Pre-processing text before topic modelling

    - Tokenize, normalize
    - Stop word removal
    - Stemming
    
Convert tokenized documents to a document - term matrix

Build LDA models on the doc-term matrix
